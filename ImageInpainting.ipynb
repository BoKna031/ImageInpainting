# Preview of images

```python
import matplotlib.pyplot as plt
import os

# Postavite putanju do direktorijuma u kojem su slike
image_directory = "mirflickr25k/mirflickr"

# Preuzmite listu svih slika
image_files = os.listdir(image_directory)

# Prikazivanje prvih 3 slike
for image_file in image_files[:3]:
    image_path = os.path.join(image_directory, image_file)
    image = plt.imread(image_path)
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```
